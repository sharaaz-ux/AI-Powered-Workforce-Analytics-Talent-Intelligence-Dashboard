Python Task
Task: Employee Attrition Analysis Dashboard
Objective

Analyze employee data and identify:

Total employees
Attrition count
Average salary
Average age
Department-wise employee count
Attrition percentage
Visualize the results




```python
import pandas as pd

# Create sample employee data
data = {
    "EmployeeID": [101, 102, 103, 104, 105],
    "Name": ["Ali", "Sara", "John", "Fatima", "Ahmed"],
    "Age": [25, 30, 28, 35, 27],
    "Department": ["IT", "HR", "Finance", "IT", "Sales"],
    "Salary": [40000, 50000, 45000, 60000, 42000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display data
print(df)

print("\nTotal Employees:", len(df))
print("Average Age:", df["Age"].mean())
print("Average Salary:", df["Salary"].mean())

print("\nEmployees in each Department:")
print(df["Department"].value_counts())
```

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect("company.db")
cursor = conn.cursor()
cursor.execute("DROP TABLE IF EXISTS Employees")
cursor.execute("""
CREATE TABLE Employees (
    EmpID INTEGER PRIMARY KEY,
    Name TEXT,
    Department TEXT,
    Salary INTEGER,
    Experience INTEGER
)
""")
employees = [
    (1, "Rahul", "IT", 60000, 2),
    (2, "Aisha", "HR", 45000, 3),
    (3, "Kiran", "IT", 70000, 5),
    (4, "Priya", "Finance", 80000, 6),
    (5, "John", "Marketing", 50000, 2),
    (6, "Sara", "HR", 55000, 4),
    (7, "Rohan", "IT", 90000, 8),
    (8, "Meera", "Finance", 65000, 3)
]

cursor.executemany("INSERT INTO Employees VALUES (?, ?, ?, ?, ?)", employees)
conn.commit()

print("\n--- All Employees Data ---")
display(pd.read_sql("SELECT * FROM Employees", conn))

print("\n--- Employees with Salary > 60000 ---")
display(pd.read_sql("""
SELECT Name, Salary
FROM Employees
WHERE Salary > 60000
""", conn))

print("\n--- Average Salary by Department ---")
display(pd.read_sql("""
SELECT Department,
AVG(Salary) AS Average_Salary
FROM Employees
GROUP BY Department
""", conn))

print("\n--- Employees with Experience > 3 years ---")
display(pd.read_sql("""
SELECT *
FROM Employees
WHERE Experience > 3
""", conn))

print("\n--- Employee with the Highest Salary ---")
display(pd.read_sql("""
SELECT *
FROM Employees
ORDER BY Salary DESC
LIMIT 1
""", conn))

cursor.execute("""
DELETE FROM Employees
WHERE EmpID = 5
""")

conn.commit()

conn.close()
```
