**Prompt : "Write a Python program using Pandas to create an employee dataset, display the data, calculate basic statistics (total employees, average age, average salary), and count employees in each department."**




```python
import pandas as pd

# Create sample employee data
data = {
    "EmployeeID": [101, 102, 103, 104, 105],
    "Name": ["Ali", "Sara", "John", "Fatima", "Ahmed"],
    "Age": [25, 30, 28, 35, 27],
    "Department": ["IT", "HR", "Finance", "IT", "Sales"],
    "Salary": [40000, 50000, 45000, 60000, 42000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display data
print(df)

print("\nTotal Employees:", len(df))
print("Average Age:", df["Age"].mean())
print("Average Salary:", df["Salary"].mean())

print("\nEmployees in each Department:")
print(df["Department"].value_counts())
```
